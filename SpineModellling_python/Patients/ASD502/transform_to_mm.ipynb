```python
import pandas as pd

px = 0.179363  # mm/pixel

df = pd.read_csv("C:\GBW_MyPrograms\Fabio_Galbusera_algorithm_CR\Fabio_Galbusera_algorithm_CR\patients\ASD-502\ASD_502_edited.txt", header=None)
df.columns = ["Vertebra","Label","X","Y","Z"]

df["Label2"] = df["Vertebra"] + "_" + df["Label"]

df["Xmm"] = df["X"] * px
df["Ymm"] = df["Y"] * px
df["Zmm"] = df["Z"] * px     

df_out = df[["Label2","Xmm","Ymm","Zmm"]]
df_out.to_csv("C:\GBW_MyPrograms\Fabio_Galbusera_algorithm_CR\Fabio_Galbusera_algorithm_CR\patients\ASD-502\ASD_502_landmarks_mm.csv", index=False)


```
